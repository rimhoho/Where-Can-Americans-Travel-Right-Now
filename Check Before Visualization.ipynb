```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_json('https://covid.ourworldindata.org/data/owid-covid-data.json').T.reset_index()
df
```

```python
# Specific country: United States
us = df.loc[lambda df:df['location'] == 'United States']['data'].tolist()
us[0][0]
```

```python
for arr in us:
    USA = pd.DataFrame({'date': us['date'],
                         'total_cases': us['aged_65_older'],
                         'new_cases': us['new_cases'],
                         'gdp_per_capita': us['gdp_per_capita'],
                         'cvd_death_rate': us['cvd_death_rate'],
                         'diabetes_prevalence': us['diabetes_prevalence']
                        }).reset_index().drop(columns=['index'])
USA.head(3)
```

```python
df_world = pd.DataFrame({'date': [pd.to_datetime(d.get('date'),format='%Y-%m-%d') for d in df.loc[210]['data']],
                      'new_cases':[d.get('new_cases') for d in df.loc[210]['data']],
                      'new_deaths':[d.get('new_deaths') for d in df.loc[210]['data']],
                      'total_cases':[d.get('total_cases') for d in df.loc[210]['data']],
                      'total_deaths':[d.get('total_deaths') for d in df.loc[210]['data']],
#                       'new_cases_per_million':[d.get('new_cases_per_million') for d in df.loc[210]['data']],
#                       'new_deaths_per_million':[d.get('new_deaths_per_million') for d in df.loc[210]['data']],
#                       'total_cases_per_million':[d.get('total_cases_per_million') for d in df.loc[210]['data']],
#                       'total_deaths_per_million':[d.get('total_deaths_per_million') for d in df.loc[210]['data']]
                     })
df_world.head(3)
```

```python
# df_world.sort_values(by=['new_cases_per_million'], inplace=True)
```

```python
# df_world_new = add_daily_measures(df_world)
df_world['Cases:14-day'] = df_world['new_cases'].rolling(14).mean()
df_world['Deaths:14-day'] = df_world['new_deaths'].rolling(14).mean()
df_world.head(16)
```

```python
df_world_new = df_world.groupby('date',as_index=False)['new_cases','new_deaths'].sum()
fig = go.Figure(data=[
    go.Bar(name='New Cases', x=df_world['date'], y=df_world['new_cases']),
    go.Bar(name='New Deaths', x=df_world['date'], y=df_world['new_deaths'])
])
fig.add_trace(go.Scatter(name='Cases:14-day',x=df_world['date'],y=df_world['Cases:14-day'],marker_color='black'))
fig.add_trace(go.Scatter(name='Deaths:14-day',x=df_world['date'],y=df_world['Deaths:14-day'],marker_color='darkred'))

# Change the bar mode
fig.update_layout(barmode='overlay', title='Worldwide daily Case and Death count',showlegend=False)
fig.show()
```
